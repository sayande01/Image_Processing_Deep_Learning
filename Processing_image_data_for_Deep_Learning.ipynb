### Resizing the image using Pillow library

```python
from PIL import Image 
```

```python
img = Image.open('golden.jpg')
img_resized = img.resize((400, 400))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the image from numpy array

img_res = mpimg.imread('dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Converting RGB images to Grayscale image using OpenCV**

```python
!pip install opencv-python
```

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('golden.jpg')
```

```python
type(img)
```

```python
img.shape
```

Converting RGB img into B/W image

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

### Saving the Grayscale Image

```python
# displaying the image
cv2.imshow('Grayscale', grayscale_image) 
```

```python
# saving the grayscale image
cv2.imwrite('golden_retriever_grayscale_image.jpg', grayscale_image)
```
